```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter 
```

```python
file = 'predicted.jpg'
first_file = cv2.imread('0.png')
first_file = cv2.cvtColor(first_file, cv2.COLOR_BGR2RGB)
```

# OpenCV

```python
img = cv2.imread(file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Pillow

```python
with Image.open(file) as img:
    img.load()
```

```python
def info(img):
    return 'Format: {}, size: {}, mode: {}'.format(img.format, img.size, img.mode)
```

```python
info(img)
new_img = img.filter(ImageFilter.SHARPEN)
```

```python
kernel = np.array([[0, -1, 0], 
                   [-1, 5, -1], 
                   [0, -1, 0]])
new_img = cv2.filter2D(img, -1, kernel)
```

```python

plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.title('Before', fontsize = 15)
plt.imshow(first_file)
plt.axis('off')
plt.subplot(1,3,2)
plt.title('Predicted', fontsize = 15)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,3)
plt.title('Processed cv2/pillow', fontsize = 15)
plt.imshow(new_img)
plt.axis('off')
plt.show()
```
